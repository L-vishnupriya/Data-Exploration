```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path = '/content/drive/MyDrive/data'
```

```python
categories = ['cherry', 'grape', 'jackfruit']

selected_images = []
dataset_dir = '/content/drive/My Drive/data'


# Iterate over categories
# Iterate over categories
for category in categories:
    category_path = os.path.join(path, category)
    print("Category Path:", category_path)  # Print category path for debugging
    if os.path.exists(category_path):  # Check if the directory exists
        images = os.listdir(category_path)
        # Take 10 random images from each category
        selected_images.extend([(os.path.join(category_path, image), category) for image in random.sample(images, 10)])
    else:
        print(f"Directory not found: {category_path}")

    # Take 10 random images from each category
    selected_images.extend([(os.path.join(category_path, image), category) for image in random.sample(images, 10)])

# Shuffle the selected images
random.shuffle(selected_images)

# Print image IDs and class names for the first 10 images
for image_path, class_name in selected_images[:10]:
    image_id = os.path.basename(image_path).split('.')[0]
    print(f"Image ID: {image_id}, Class Name: {class_name}")

# Display the selected and shuffled images one after the other
plt.figure(figsize=(15, 6))
for idx, (image_path, class_name) in enumerate(selected_images[:10]):
    img = Image.open(image_path)

    plt.subplot(2, 5, idx + 1)
    plt.imshow(img)
    plt.title(class_name)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
